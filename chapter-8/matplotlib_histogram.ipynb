```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


np.random.seed(45)
data = pd.DataFrame(np.random.randint(0,101,1000), columns=['Scores'])

data.hist(bins=20, color='lightblue', edgecolor='black')

plt.title('Distribution of Exam Scores')
plt.xlabel('Score')
plt.ylabel('Number of Students')
plt.show()
```
